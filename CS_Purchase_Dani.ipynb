**Statistic Analysis**

1. Data collection

2. Preparing samples

3. Creating two hypothesis (null hypothesis, alternate hypothesis)

4. Apply appropriate test (based on requirement or case)

5. p_value>0.05 -> accept null hypothesis

6. p_value<0.05 -> reject null hypothesis

It was observed that the average purchase rate made by the Men of the age 18-25 was 10000. Is it still the same?

```python
new_data=data[(data['Age']==1)&data['Gender']==1]
new_data
```

```python
new_data['Purchase'].mean()
```

```python
#random_state means seedability
```

```python
sample=new_data.sample(3600,random_state=5)
```

```python
sample
```

```python
#null hypothesis = mean will be equal to 10000
#alternate hypothesis = mean will not be equal to 10000
```

```python
#one sample t-test
from scipy.stats import ttest_1samp
```

```python
a_mean=10000
```

```python
t_stat,p_value=ttest_1samp(sample['Purchase'],a_mean)
print(t_stat,p_value)
```

```python
if p_value>0.05:
  print('null hypothesis = mean will be equal to 10000')
else:
  print('alternate hypothesis = mean will not be equal to 10000')
```

2. It was obeserved that the percentage of women that spend more than 10000 was 35%. Is it still the same?

```python
#null hypothesis = proportion is 35%
#alternate hypothesis = proportion is not 35%
```

```python
data_new=data[(data['Purchase']>10000)]
```

```python
data_new.shape
```

```python
data_new['Gender'].value_counts()
```

```python
18685/91057
```

```python
#z_proportion test
```

```python
#No of womens in the sample
count=data_new['Gender'].value_counts()[0]

#number of observation
nobs=len(data_new['Gender'])

#hypothesised value
p0=0.35
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
z_stat,p_val=proportions_ztest(count=count,nobs=nobs,value=p0)
print(z_stat,p_val)
```

```python
if p_value>0.05:
  print('null hypothesis = proportion is 35%')
else:
  print('alternate hypothesis = proportion is not 35%')
```
